```python
def handle_outliers(column):
    Q1 = train[column].quantile(0.25)
    Q3 = train[column].quantile(0.75)
    IQR = Q3 - Q1
    print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")

    lbound = Q1 - 1.5 * IQR
    ubound = Q3 + 1.5 * IQR
    print(f"Lower Bound: {lbound}, Upper Bound: {ubound}")

    train[column] = np.where(train[column] > ubound, ubound,
                          np.where(train[column] < lbound, lbound, train[column]))

for col in train.columns:
    if pd.api.types.is_numeric_dtype(train[col]) and not pd.api.types.is_bool_dtype(train[col]):
        handle_outliers(col)
        sns.boxplot(train[col])
        plt.title(f"Boxplot of {col}")
        plt.show()
```

```python
def handle_outliers(column):
    Q1 = test[column].quantile(0.25)
    Q3 = test[column].quantile(0.75)
    IQR = Q3 - Q1
    print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")

    lbound = Q1 - 1.5 * IQR
    ubound = Q3 + 1.5 * IQR
    print(f"Lower Bound: {lbound}, Upper Bound: {ubound}")

    test[column] = np.where(test[column] > ubound, ubound,
                          np.where(test[column] < lbound, lbound, test[column]))

for col in test.columns:
    if pd.api.types.is_numeric_dtype(test[col]) and not pd.api.types.is_bool_dtype(train[col]):
        handle_outliers(col)
        sns.boxplot(x=test[col])
        plt.title(f"Boxplot of {col}")
        plt.show()
```

```python
x = train.drop(columns = "Transported")
y = train["Transported"]
x
```

```python
label = LabelEncoder()
for col in x.columns:
    if pd.api.types.is_object_dtype(x[col]) or pd.api.types.is_bool_dtype(x[col]):
        x[col] = label.fit_transform(x[col])
```

```python
label = LabelEncoder()
for col in test.columns:
    if pd.api.types.is_object_dtype(test[col]) or pd.api.types.is_bool_dtype(test[col]):
        test[col] = label.fit_transform(test[col])
```

```python
from sklearn.feature_selection import chi2
chi = chi2(x, y)
chiv = pd.Series(chi[0], index=x.columns)  # Chi-squared statistics
p_values = pd.Series(chi[1], index=x.columns)  # P-values

# Sort based on chi-squared statistics
chiv.sort_values(inplace=True)
chiv.plot.bar()

# Filter out "unwanted" columns based on p-values
unwanted_cols = p_values[p_values > 0.05].index
print(f"Unwanted columns based on p-values: {unwanted_cols}")

chiv
```

```python
y.value_counts()/len(train)     # so the probablity is almost imbalanced
```

```python
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
test = scaler.fit_transform(test)
```

```python

from sklearn.ensemble import GradientBoostingClassifier, BaggingClassifier

# Initialize the Gradient Boosting Classifier
gb_model = GradientBoostingClassifier()

# Initialize the Bagging Classifier with Gradient Boosting as the base estimator
bagging_model = BaggingClassifier(estimator=gb_model, n_estimators=10, random_state=42)

# Fit the Bagging model on the training data
bagging_model.fit(x, y)

pred = bagging_model.predict(test)
```

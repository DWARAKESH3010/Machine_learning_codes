```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'Height': [175, 160, 165, 180, 155, 170]
})

# Calculate the sum of heights for each gender
sum_height_per_gender = data.groupby('Gender')['Height'].sum()

# Plot the bar plot
plt.figure(figsize=(10, 6))
plt.bar(sum_height_per_gender.index, sum_height_per_gender.values, color='skyblue')
plt.xlabel('Gender')
plt.ylabel('Total Height')
plt.title('Total Height by Gender')
plt.show()
```

```python
high = data["Age"].quantile(0.75)
low= data["Age"].quantile(0.25)
print(high,low)
IQR = high - low
upper = high  +1.5*IQR
lower = low  - 1.5*IQR
print(upper,lower)
```

```python
def limit(value):
    if value > upper:
        return upper
    elif value < lower:
        return lower
    else:
        return value
```

```python
data["Age"] = data["Age"].apply(limit)
```

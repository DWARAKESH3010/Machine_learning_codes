```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv("Amazon Sale Report - Amazon Sale Report.csv.csv")
data = df.copy()
data.drop(["New","PendingS"], axis = 1, inplace = True)
data.head()
```

```python
data["Amount"].describe()
```

# amount handle missing value

```python
data['Amount'].fillna(data['Amount'].median(), inplace=True)
# Assuming 'data' is your DataFrame
data['currency'] = data['currency'].fillna('INR')
data['ship-country'] = data['ship-country'].fillna('IN')
# Display the first 30 rows of the 'currency' column

```

# outliers

```python
high = data["Amount"].quantile(0.75)
low= data["Amount"].quantile(0.25)
IQR = high - low
upper = high  +1.5*IQR
lower = low  - 1.5*IQR
def limit(value):
    if value > upper:
        return upper
    elif value < lower:
        return lower
    else:
        return value
    
data["Amount"] = data["Amount"].apply(limit)

sns.boxplot(data["Amount"])
plt.title("Amount")
```

```python

# Assuming 'data' is your DataFrame
numeric_columns = data.select_dtypes(include=['int64', 'float64']).columns

# Calculate the correlation matrix
correlation_matrix = data[numeric_columns].corr()

# Display the correlation coefficients
print(correlation_matrix)
```
